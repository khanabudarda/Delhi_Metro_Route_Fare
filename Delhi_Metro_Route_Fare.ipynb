```python
#Importing all Necessary Libraries
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import os
from tqdm import tqdm_notebook
#Initalizing the Driver
driver = webdriver.Chrome()
#Getting the Response from DMRC site
response=driver.get("http://www.delhimetrorail.com/metro-fares.aspx?fs=SAQY8wVbPA8lld&ts=s045vnclU5TvIlld")
for i in tqdm_notebook(range(67,len(station_lst3))):
    for j in tqdm_notebook(range(len(station_lst3))):
        to = driver.find_element_by_id("ctl00_MainContent_ddlFrom")
        fro = driver.find_element_by_id("ctl00_MainContent_ddlTo")
        to.send_keys(station_lst3[i])
        fro.send_keys(station_lst3[j])
        driver.find_element_by_id("ctl00_MainContent_btnShowFare").click()
        route=driver.find_element_by_class_name("fr_stations")
        route2=route.text
        routef=route2.split('\n')
        fare = driver.find_element_by_class_name("fare_new_nor_right")
        faref=fare.text
        fare_lst.append(faref)
        to_lst.append(station_lst3[i])
        from_lst.append(station_lst3[j])
        route_list.append(routef)
```
